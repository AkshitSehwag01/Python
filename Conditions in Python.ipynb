#### 5.2. To check if a given number (between 0 and 100) meets the following conditions

Divisible by 5, 7, or 9 (or a combination of them).
Print which numbers it is divisible by and which it is not.
If the number is out of range, display a corresponding message.
Handle cases where the number is divisible by all three."

```python
num = int(input("Enter a number: "))

if 0 <= num <= 100:
    # Store divisibility checks in variables for better performance
    div5 = (num % 5 == 0)
    div7 = (num % 7 == 0)
    div9 = (num % 9 == 0)

    if div5 and not div7 and not div9:
        print("5-5: Divisible by 5 but NOT by 7 and 9.")
    elif div7 and not div5 and not div9:
        print("7-7: Divisible by 7 but NOT by 5 and 9.")
    elif div9 and not div5 and not div7:
        print("9-9: Divisible by 9 but NOT by 5 and 7.")
    elif div5 and div7 and not div9:
        print("(5,7): Divisible by 5 and 7 but NOT by 9.")
    elif div5 and div9 and not div7:
        print("(5,9): Divisible by 5 and 9 but NOT by 7.")
    elif div7 and div9 and not div5:
        print("(7,9): Divisible by 7 and 9 but NOT by 5.")
    elif div5 and div7 and div9:
        print("(5,7,9): Divisible by 5, 7, and 9.")
    else:
        print("The number does not meet any condition.")

else:
    print("The number is OUT of the range (0-100).")
```
